```python
#Plotting a bar graph to show the minimum number of coin flips needed to get at least one head with 90% probability

import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

target_prob = 0.9
p = 0.5

n = 1
while 1 - p**n <= target_prob:
    n += 1

print(f"Minimum number of flips needed to have at least 90% chance of one head: {n}")

trials = 10000
results = []

for flips in range(1, n+5): 
    success_count = 0
    for _ in range(trials):
        outcome = np.random.choice(['H', 'T'], size=flips)
        if 'H' in outcome:
            success_count += 1
    prob = success_count / trials
    results.append((flips, prob))

flips, probs = zip(*results)
sns.set(style='whitegrid')
sns.lineplot(x=flips, y=probs, marker='o')
plt.axhline(y=0.9, color='red', linestyle='--', label='90% Threshold')
plt.xlabel('Number of flips')
plt.ylabel('Probability of at least one Head')
plt.legend()
plt.show()
```
